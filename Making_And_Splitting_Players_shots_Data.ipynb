### Merge all bundesliga Players CSVs into one

```python
import os
import pandas as pd

# Folder containing CSV files
folder_path = r'C:\Users\Youcode\Desktop\8 Months\Fil Rouge\datasets\bundesliga\shots'

# List all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Iterate through each CSV file, add a 'season' column, and save it
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    # Extract the league from the file name
    league = file.split('_')[1]
    # # Extract the year from the file name
    # year = file.split('_')[-1].split('.')[0]
    # start_year = int(year[:2]) + 2000
    # end_year = int(year[-2:]) + 2000
    # year_range = f"{start_year}-{end_year}"
    # Read the CSV file
    df = pd.read_csv(file_path)
    # Add a 'league' column with the corresponding league
    df['league'] = league
    # Add a 'season' column with the corresponding year range
    # df['season'] = year_range
    # Save the updated CSV file
    df.to_csv(file_path, index=False)

# Merge all CSVs into one
all_csv_data = [pd.read_csv(os.path.join(folder_path, file)) for file in csv_files]

# Concatenate all dataframes
shots_bundesliga_df = pd.concat(all_csv_data, ignore_index=True)

# Save the merged dataframe to a single CSV file
shots_bundesliga_df.to_csv(os.path.join(r'C:\Users\Youcode\Desktop\8 Months\Fil Rouge\datasets\bundesliga', 'shots_bundesliga_dataset.csv'), index=False)
```

# Merge all leagues

```python
import os
import pandas as pd

# Define the folder path where the league folders are located
folder_path = r'C:\Users\Youcode\Desktop\8 Months\Fil Rouge\datasets'

# List all folders (leagues) in the datasets folder
leagues_folders = [f for f in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, f))]

# Create an empty list to store DataFrames of each league
dataframes = []

# Iterate through each league folder and read CSV files
for league in leagues_folders:
    league_folder_path = os.path.join(folder_path, league)
    csv_files = [file for file in os.listdir(league_folder_path) if file.endswith('.csv')]
    
    # Read each CSV file and append it to the list of DataFrames
    for file in csv_files:
        file_path = os.path.join(league_folder_path, file)
        df = pd.read_csv(file_path)
        dataframes.append(df)

# Concatenate all DataFrames into a single DataFrame
merged_df = pd.concat(dataframes, ignore_index=True)

# Save the merged DataFrame to a new CSV file
merged_df.to_csv('shots_of_leagues_dataset.csv', index=False)
```

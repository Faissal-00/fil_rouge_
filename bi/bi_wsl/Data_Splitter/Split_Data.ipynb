```python
import pandas as pd

data=pd.read_csv('players_of_leagues_dataset.csv')
data.head()
```

```python
# Split the data based on the 'league' column
csv_data = data[data['league'].isin(['bundesliga', 'epl', 'la_liga'])]
json_data = data[data['league'] == 'ligue_1']
api_data = data[data['league'] == 'serie_a']

# Save the split data to respective files
csv_data.to_csv('datasets/csv_players.csv', index=False)
json_data.to_json('datasets/json_players.json', orient='records', lines=True)
api_data.to_csv('datasets/api_players.csv', index=False)
```
